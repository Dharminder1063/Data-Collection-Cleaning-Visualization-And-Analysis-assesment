Input:

```python
a = np.array([2, 6, 1, 9, 10, 3, 27])

items = a[(a >= 5) & (a <= 10)]

print(items)
```

10. Create an array of numbers between 0 and 6 with increments of 0.3 and name its “x”.
Then on the same plot, plot x, x², x³, and x⁴. For consistency, use the following style lines
respectively, “ro”, “bs”, “g”, and “:”. Lastly, make sure that the x-axis covers 0 to 6, while
the y-axis spans from 0 to 125. Do not worry if you are not familiar with the style lines —
you will recognize them as soon as you see the plot.

```python
x = np.arange(0, 6.1, 0.3)

fig, ax = plt.subplots()

ax.plot(x, x, 'ro', label='x')     # Plot y=x
ax.plot(x, x**2, 'bs', label='x²')   # Plot y=x^2
ax.plot(x, x**3, 'g', label='x³')    # Plot y=x^3
ax.plot(x, x**4, ':', label='x⁴')   # Plot y=x^4

ax.set_xlim(0, 6)
ax.set_ylim(0, 125)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Various Power Functions")
ax.legend()

plt.show()
```

12. Plot a histogram of x, where x consists of 100,000 randomly-selected points with a normal
distribution (hint: you can use numpy.random.randn() to generate the random points). The
histogram should have 10 bins. Look at how the histogram changes when we try 20 and 50
bins.

```python
x = np.random.randn(100000)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(x, bins=10, color='blue', edgecolor='black')
axes[0].set_title("Histogram with 10 bins")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].grid(axis='y', alpha=0.75)

axes[1].hist(x, bins=20, color='green', edgecolor='black')
axes[1].set_title("Histogram with 20 bins")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")
axes[1].grid(axis='y', alpha=0.75)

axes[2].hist(x, bins=50, color='red', edgecolor='black')
axes[2].set_title("Histogram with 50 bins")
axes[2].set_xlabel("Value")
axes[2].set_ylabel("Frequency")
axes[2].grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()
```

14. List a few limitations of TensorFlow.

1. Steep Learning Curve in its Early Versions
TensorFlow 1.x was notorious for its complexity. It required users to manually define static computational graphs, which was counter-intuitive for many Python developers used to imperative programming. While TensorFlow 2.x and Keras integration have significantly eased this (through eager execution), the sheer size and depth of its API still present a steeper learning curve compared to more minimalistic frameworks like PyTorch.
2. Large Overhead and Debugging Challenges in Production
Although TensorFlow excels in deployment tools (like TensorFlow Serving and Lite), the process of converting models to optimized graph formats for deployment (e.g., TFLite, SavedModel) can sometimes be complex. Debugging issues that arise specifically within these optimized, production-level graph modes can be more challenging than in a pure execution environment.
3. Verbosity of Low-Level APIs
While Keras provides a high-level abstraction, using the core TensorFlow APIs (especially for highly custom operations or advanced research where Keras might be limiting) can be verbose and require more lines of code compared to other frameworks.
4. Significant Ecosystem Complexity
TensorFlow is an ecosystem of many tools (TensorBoard, TFX, TF Serving, TF Lite, TF Hub, etc.). While this offers a comprehensive MLOps solution, managing and integrating all these different components can introduce operational complexity.
5. Performance Issues with Eager Execution (Historically)
Eager execution (the default in TF2.x) makes debugging easier, but it can be slower than static graph execution for large-scale production training. Developers must use @tf.function decorators to convert eager code into graphs to achieve peak performance, adding an extra layer of complexity to performance optimization.
6. Resource Intensive
TensorFlow applications can sometimes have a large memory footprint and require significant computational resources, especially when managing extensive graphs and large datasets, which might be a limitation in low-resource environments outside of optimized TFLite deployments.

15. What do you know about supervised and unsupervised machine learning?

Supervised and unsupervised learning are the two primary categories of machine learning, fundamentally differing in their approach to data and the presence of "labels" or "answers" within that data.

Supervised Machine Learning

Supervised learning involves training a model on a labeled dataset. This means the input data (features) has a corresponding output (target or label) that the model learns to predict. It's like learning with a teacher or a supervisor.

Unsupervised Machine Learning

Unsupervised learning involves training a model on data that does not have labeled outputs. In this scenario, the model must discover hidden patterns, structures, or relationships within the data entirely on its own. It's like learning without a teacher.
